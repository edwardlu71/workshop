### TanH

```python
f(x) = tanh(x)
df = f.derivative()
#df = 1 - f(x)^2
df
```

```python
plot(f, -10, 10)
```

```python
plot(df, -10, 10)
```

### ReLU

```python
f(x) = piecewise([((-1000,0), 0), ([0, 1000], x)]);  f
```

```python
plot(f, -10, 10)
```

### Leaky ReLU

```python
f(x) = piecewise([((-1000,0), 0.01*x), ([0, 1000], x)]);  f
```

```python
plot(f, -10, 10)
```

## ELu (Edward Lu?)

good thing for Elu is
- it is smooth
- converge cost to zero faster and produce more accurate results

```python
a = 1
f(x) = piecewise([((-1000,0), a * (exp(x)-1)), ([0, 1000], x)]);  f
```

```python
plot(f, -10, 10)
```

### Step Function

```python
step(x) = piecewise([((-1000,0), 0), ([0, 1000], 1)]);  step
```

```python
plot(step, -10, 10)
```

```python
var('x')
sigmoid = 1/(1+exp(-x))
g = Graphics()
g += plot(1/(1+exp(-x)), -10, 10, legend_label=' Sigmoid', thickness=5, color='red')
g += plot(piecewise([((-1000,0), 0.01*x), ([0, 1000], x)]), -10, 1, legend_label=' Leaky ReLU', thickness=5, color='orange')
g += plot(piecewise([((-1000,0), 0), ([0, 1000], 1)]), -10, 10, legend_label=' Step', thickness=5, color='blue')
g += plot(tanh(x), -10, 10, legend_label=' TanH', thickness=5, color='green')
g.show()
```
