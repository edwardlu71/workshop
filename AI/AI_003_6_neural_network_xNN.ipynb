### RNN: Recurrent Neural Network

Recurrent neural networks are networks that are designed to interpret temporal or sequential information. RNNs use other data points in a sequence to make better predictions. They do this by taking in input and reusing the activations of previous nodes or later nodes in the sequence to influence the output.




Applied to:

- translation
- story telling
- autocoding
- music


### LSTM: Long Short Term Memory

LSTM is the most popular RNN architecture. It has feedback connections. It can not only process single data points (such as images), but also entire sequences of data (such as speech or video)

RNN's problem: because of backpropagation on long sequence chain of the events

- gradient vanishing (w<1)
- gradient exploding (w>1)

To solve the problem, the classic LSTM architecture is characterized by a **persistent linear cell state** surrounded by non-linear layers feeding input and parsing output from it. Concretely the cell state works in concert with 4 gating layers, these are often called the forget, (2x) input, and output gates.



Applied to

- classifying
- trading, **prediction on time series data** (working on it, candidate for next topic)
